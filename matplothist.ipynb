### Basic Graph

```python
X = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

plt.plot(
    X,
    y,
    color="#468000",
    label="2 times X",
    marker="D",
    markersize=10,
    markeredgecolor="blue",
    markerfacecolor="red",
    linewidth=5,
)
plt.title("xOur First Graph", fontdict={"fontsize": 15, "fontname": "Arial"})
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks([0, 1, 2, 3, 4, 5])
plt.yticks(range(0, 9, 2))
plt.legend()

plt.show()
```

```python
X = [0, 1, 2, 3, 5]
y = [0, 2, 4, 6, 9]
# color , maker , style
plt.plot(X, y, "r*--", label="2 times X")
plt.title("xOur First Graph", fontdict={"fontsize": 15, "fontname": "Arial"})
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks([0, 1, 2, 3, 4, 5, 6])
plt.yticks(range(0, 9, 2))
plt.legend()

plt.show()
```

```python
X = [0, 1, 2, 3, 5]
y = [0, 2, 4, 6, 9]

plt.figure(figsize=(6, 4), dpi=150)

# color , maker , style
plt.plot(X, y, "r*--", label="2 times X")
a = np.arange(0, 5)
plt.plot(a[:3], a[:3] ** 2, "c", label="X power 2")
plt.plot(a[2:], a[2:] ** 2, "b--", label="X power 2")


plt.title("Our First Graph", fontdict={"fontsize": 15, "fontname": "Arial"})
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks([0, 1, 2, 3, 4])
plt.yticks(range(0, 25, 2))
plt.legend()

plt.savefig("Graphyy", dpi=300)

plt.show()
```

```python
def strip(X):
    return int(X.strip("lbs"))
```

```python
weight = fifa.Weight.dropna().map(lambda X: int(X.strip("lbs")))
```

```python
light = sum(weight.map(lambda x: x in range(0, 125)))
ligtt_medium = sum(weight.map(lambda x: x in range(125, 150)))
medium = sum(weight.map(lambda x: x in range(150, 175)))
medium_heavy = sum(weight.map(lambda x: x in range(175, 200)))
heavy = sum(weight.map(lambda x: x in range(200, 400)))
```

```python
light
```

```python
labels = ["light", "ligtt_medium", "medium", "medium_heavy", "heavy"]
plt.figure(figsize=(5, 5))
plt.pie(
    [light, ligtt_medium, medium, medium_heavy, heavy],
    labels=labels,
    autopct="%0.2f",
    pctdistance=0.7,
    explode=(0.5, 0.5, 0, 0, 0.5),
)
plt.show()
```

### BOX PLOT

```python
clubs = fifa.Club[:5]
plt.figure(figsize=(15, 8))
for i, club in enumerate(clubs):
    plt.subplot(2, 3, i + 1)
    plt.boxplot(fifa[fifa.Club == club]["Overall"], labels=[club])
```

```python
overalls = []
for i in clubs:
    overalls.append(fifa[fifa.Club == i]["Overall"])
plt.figure(dpi=150)
boxes = plt.boxplot(overalls, labels=clubs.values, patch_artist=True)
plt.xticks(size=6)
for box in boxes["boxes"]:
    box.set(color="b", linewidth=1.2)
    box.set(facecolor="#819ad6")
    # box.set(medcol="b")
plt.show()
```

```python
import seaborn as sns
```

```python
def get_clause(x):
    x = str(x).strip("")
    if "M" in x:
        x = float(x.strip("M")) * 1000
    elif "K" in x:
        x = x.strip("K")
    return float(x)
```

```python
fifa["Numeric Clause"] = fifa["Release Clause"].map(get_clause)
```

```python
plt.figure(dpi=100)
sns.lmplot(y="Numeric Clause", x="Overall", data=fifa)
```

```python
plt.figure(dpi=100)
sns.lmplot(
    y="Numeric Clause", x="Overall", data=fifa, fit_reg=False, hue="Preferred Foot"
)
plt.ylim(50000, None)
plt.xlim(80, 93)
plt.show()
```

# Violin plot

```python
plt.figure(dpi=100)
first_five_position = fifa[
    (fifa.Position.isin(fifa.Position.unique()[15:18]))
    & (fifa["Numeric Clause"] < 1000)
]

sns.violinplot(
    x="Position",
    y="Numeric Clause",
    data=first_five_position,
    palette=["r", "b", "y", "g", "c"],
)
```

# Swarmplot plot 

### swarmplot for seeing the distribution in dots (              )

```python
plt.figure(dpi=100)
sns.swarmplot(
    x="Position",
    y="Numeric Clause",
    data=first_five_position,
    palette=["r", "b", "y", "g", "c"],
)# Violin plot
```
